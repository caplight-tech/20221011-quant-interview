# Read the data

```python
path_to_file = "/Users/olgaponomarenko/Downloads/deltaneutral-vol-data_L2_20220824/options_20220824.csv"
data = pd.read_csv(path_to_file)

# For simplicity, let's only analyze call options
data = data.loc[data["Type"] == "call"]

# Columns
data.loc[:, "moneyness"] = data["Strike"] / data["UnderlyingPrice"]
data.loc[:, "implied_vol"] = data["IV"]
data.loc[:, "maturity"] = (pd.to_datetime(data["Expiration"]) - pd.to_datetime(data[" DataDate"])).dt.days / 365
data.loc[:, "ticker"] = data["UnderlyingSymbol"]
data = data[["ticker", "moneyness", "maturity", "implied_vol"]].copy() # we don't need other columns for this exercise
```

# Visualize

```python
tesla_vol = data.loc[data["ticker"] == "TSLA"]

fig = plt.figure(figsize=(6,8), dpi=100)
ax = plt.axes(projection='3d')

ax.scatter3D(tesla_vol["moneyness"], tesla_vol["maturity"], tesla_vol["implied_vol"], c=tesla_vol["implied_vol"]/2, cmap='Reds')
ax.set_xlabel("Moneyness")
ax.set_ylabel("Maturity")
ax.set_zlabel("Implied vol")
```
